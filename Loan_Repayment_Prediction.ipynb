```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree

balance_data=pd.read_csv('C:/Users/kadam/Downloads/loan_repayment_data.csv',sep=',',header=0)
```

```python
print('DataLenght: ',len(balance_data))
print('Dataset Shape: ',balance_data.shape)
```

```python
print("Dataset:")
balance_data.head()
```

```python
balance_data['initial_payment']=balance_data['initial_payment'].astype(int)
balance_data['last_payment']=balance_data['last_payment'].astype(int)
```

```python
balance_data.head()
```

```python
X=balance_data.values[:,1:5]
Y=balance_data.values[:,0]
```

```python
print('X:',X.shape)
print('Y:',Y.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=100)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
clf_entropy=DecisionTreeClassifier(criterion="entropy",random_state=100,
                                   max_depth=3,min_samples_leaf=5)
clf_entropy.fit(X_train,y_train)
```

```python
y_pred_en=clf_entropy.predict(X_test)
y_pred_en
```

```python
print("Accuray is ",accuracy_score(y_test,y_pred_en)*100)
```
